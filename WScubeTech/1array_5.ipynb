```python
#Addition and removing element in the array
```

```python
import numpy as np

a = np.array([20,30,60,80])
print(np.append(a,90))
```

```python
import numpy as np

a = np.array([[20,30],[60,80]]) # in 2d array
print(np.append(a,90))
```

```python
import numpy as np

a = np.array([[20,30],[60,80]]) # in 2d array
print(np.append(a,[90,100]))
```

```python
import numpy as np

a = np.array([20,30,60,80])
print(np.insert(a,1,50))
```
