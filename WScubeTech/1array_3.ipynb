```python
#mathematical function anf function on array
import numpy as np

arr1 = np.array([30,40,60,20,10])
arr2 = np.array([30,20,10,20,20])
print(arr1+arr2)
print(np.add(arr1,arr2))
```

```python
import numpy as np

arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[10,20]])
print(arr1+arr2)
print(np.add(arr1,arr2))
```

```python
import numpy as np

arr1 = np.array([[30,40],[20,10]])
arr2 = np.array([[30,20],[10,20]])
print(arr1/arr2)
print(np.divide(arr1,arr2))
```

```python
arr1 = np.array([3,4,2,1])
arr2 = np.array(3)
print(np.power(arr1,arr2))
```

```python
arr1 = np.array([9,16,4,1])
print(np.sqrt(arr1))
```
