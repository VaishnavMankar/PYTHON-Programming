```python
import numpy as np

a = np.array([20,40,60,70])
print(np.sum(a))
print(np.min(a))
print(np.max(a))
print(np.size(a))
print(np.mean(a))
print(np.cumsum(a)) #this means adding the current number to previous number
```

```python
a = [100,150,199,200,250,130]
b = [10,50,30,40,30,10]

price = np.array(a)
quantity = np.array(b)
print(price,"\n",quantity)
print()
print(np.cumprod([price,quantity],axis = 0))

c = np.cumprod([price,quantity],axis = 0)
print("The sum of the total elements are ", c[1].sum())


```
