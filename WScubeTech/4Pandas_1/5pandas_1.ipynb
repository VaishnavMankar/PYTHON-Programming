```python
#Group By in Pandas
```

```python
import pandas as pd

data = pd.read_excel("E:/Scientific computing/project/ESD.xlsx")
print(data.head(10))
```

```python
gp = data.groupby("Department").agg({"Gender":"count"})
print(gp)
```

```python
gp = data.groupby("Job Title").agg({"EEID":"count"})
print(gp)
```

```python
gp = data.groupby(["Department","Gender"]).agg({"EEID":"count"})
print(gp)
```

```python
 gp1 = data.groupby("Country").agg({"Age":"mean"})
print(gp1)
```

```python
 gp2 = data.groupby("Country").agg({"Annual Salary":"mean"})
print(gp2)   #average salary
```

```python
 gp3 = data.groupby("Country").agg({"Annual Salary":"max"})
print(gp3)
```

```python
 gp3 = data.groupby(["Country","Gender"]).agg({"Annual Salary":"max"})
print(gp3)
```

```python
 gp3 = data.groupby(["Country","Gender"]).agg({"Annual Salary":"max","Annual Salary":"min"})
print(gp3)
```

```python
 gp3 = data.groupby(["Country","Gender"]).agg({"Annual Salary":"max","Age":"min"})
print(gp3)
```
