```python
import pandas as pd
data = {"Months":["January","February","March","April"]}

a = pd.DataFrame(data)
print(a)

def extract(value):  #returns the first three characters of that string
    return value[0:3]

a["Short_months"] = a["Months"].map(extract)
print(a)
```
