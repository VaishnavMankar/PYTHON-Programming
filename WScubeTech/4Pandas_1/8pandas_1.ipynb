```python
import pandas as pd

data = {"Keys":["k1","k2","k1","k2"],
        "Name":["John","Ben","David","Peter"],
        "Houses":["red","blue","green","red"],
        "Grades":["3rd","8th","9th","8th"]}

df = pd.DataFrame(data)
print(df)

pivot_df = df.pivot(index="Keys",columns="Name",values=["Houses","Grades"])
print(pivot_df)
```

```python
dict = {"Name":["John","Ben","David","Peter"],
        "Houses":["red","blue","green","red"],
        "Grades":["3rd","8th","9th","8th"]}
df = pd.DataFrame(dict)
print(df)

print(pd.melt(df, id_vars=["Name"],value_vars=["Houses","Grades"],var_name="House&Grades",value_name="values"))
```
