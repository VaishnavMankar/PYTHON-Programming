### 1. Counting VOWEL in a given word

```python
vowel = ['a','e','i','o','u']
word = "programming"
count = 0
for char in word:
    if char in vowel:
        count += 1

print(count)
```

### 2. Counting CONSONANTS in a given word

```python
vowel = ['a','e','i','o','u']
word = "programming"
count = 0
for char in word:
    if char not in vowel:
        count += 1

print(count)
```

### 3. Write FIBONACCI Series

```python
def fibonacci(n):
    fib = [0,1]
    for i in range(n):
        fib.append(fib[-1] + fib[-2])
    return fib

print(fibonacci(10))
```

### 4. Find the maximum number in the list

```python
numberList = [12,3,44,56,76,54,32]
max = 0
for i in numberList:
    if i > max:
        max = i
print(max)
```

### 5. Find the minimum in the list

```python
numberList = [12,3,44,56,76,54,32]
min = numberList[0]
for i in numberList:
    if i < min:
        min = i
print(min)
```

### 9. Printing the pyramid

```python
def pyramid(n):
    for i in range(n):
        for j in range(i,n):
            print(" ", end = "")
        for j in range(i+1):
            print("*",end="")
        for j in range(i):
            print("*",end="")
        print("")

pyramid(9)
```
