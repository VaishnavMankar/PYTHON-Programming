# Decision Tree for Flat Price Prediction (Without Libraries)

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the dataset
data = pd.read_csv('cleandataPune.csv')

# Check for missing values
print("Missing values in each column:")
print(data.isnull().sum())

# Optionally handle missing values
data.fillna(data.mean(), inplace=True)  # Fill with mean

# Alternatively, you could drop rows with any missing values
# data.dropna(inplace=True)

# Print the unique site_location names in list format
site_locations = data['site_location'].unique().tolist()
print("Site Locations:", site_locations)
```

```python
# Preprocess the data
# Convert 'site_location' to numerical values for modeling purposes
data['site_location'] = data['site_location'].astype('category').cat.codes

# Select features and target variable
features = data[['total_sqft', 'bath', 'site_location']].values  # Convert to numpy array
target = data['price'].values  # Convert to numpy array

# Split the data into training and testing sets (80% train, 20% test)
np.random.seed(42)  # For reproducibility
indices = np.random.permutation(len(features))
train_size = int(0.8 * len(features))

X_train = features[indices[:train_size]]
y_train = target[indices[:train_size]]
X_test = features[indices[train_size:]]
y_test = target[indices[train_size:]]
```

```python
# Decision Tree Implementation
class DecisionTree:
    def __init__(self, max_depth=None):
        self.max_depth = max_depth
        self.tree = None

    class Node:
        def __init__(self, feature=None, threshold=None, left=None, right=None, value=None):
            self.feature = feature  # Feature index for splitting
            self.threshold = threshold  # Threshold value for splitting
            self.left = left  # Left child node
            self.right = right  # Right child node
            self.value = value  # Leaf node value

    def fit(self, X, y):
        self.tree = self._grow_tree(X, y)

    def _grow_tree(self, X, y, depth=0):
        num_samples, num_features = X.shape
        
        # Check if we should stop splitting
        if num_samples == 0 or (self.max_depth is not None and depth >= self.max_depth):
            return self.Node(value=np.mean(y))  # Return leaf node with mean value
        
        # Find the best split
        best_feature, best_threshold = self._best_split(X, y)
        
        if best_feature is None:
            return self.Node(value=np.mean(y))  # Return leaf node with mean value
        
        # Split the dataset
        left_indices = X[:, best_feature] < best_threshold
        right_indices = X[:, best_feature] >= best_threshold
        
        left_node = self._grow_tree(X[left_indices], y[left_indices], depth + 1)
        right_node = self._grow_tree(X[right_indices], y[right_indices], depth + 1)
        
        return self.Node(feature=best_feature, threshold=best_threshold, left=left_node, right=right_node)

    def _best_split(self, X, y):
        best_gain = -np.inf
        best_feature = None
        best_threshold = None
        
        for feature in range(X.shape[1]):
            thresholds = np.unique(X[:, feature])
            for threshold in thresholds:
                gain = self._information_gain(X[:, feature], y, threshold)
                if gain > best_gain:
                    best_gain = gain
                    best_feature = feature
                    best_threshold = threshold
                    
        return best_feature, best_threshold

    def _information_gain(self, feature_column, y, threshold):
        parent_loss = np.mean((y - np.mean(y)) ** 2)  # Variance of parent node
        
        left_indices = feature_column < threshold
        right_indices = feature_column >= threshold
        
        if len(y[left_indices]) == 0 or len(y[right_indices]) == 0:
            return 0
        
        left_loss = np.mean((y[left_indices] - np.mean(y[left_indices])) ** 2)
        right_loss = np.mean((y[right_indices] - np.mean(y[right_indices])) ** 2)
        
        child_loss = (len(y[left_indices]) * left_loss + len(y[right_indices]) * right_loss) / len(y)
        
        return parent_loss - child_loss

    def predict(self, X):
        return np.array([self._predict(sample) for sample in X])

    def _predict(self, sample):
        node = self.tree
        while node.value is None:  # Not a leaf node
            if sample[node.feature] < node.threshold:
                node = node.left
            else:
                node = node.right
        return node.value
```

```python
# Train the Decision Tree model
model = DecisionTree(max_depth=5)  
model.fit(X_train, y_train)

# Evaluate the model on test data
predictions = model.predict(X_test)

# Calculate accuracy (Mean Absolute Percentage Error)
mape = np.mean(np.abs((y_test - predictions) / y_test)) * 100

# Print accuracy and number of leaf nodes in the tree
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")

def count_leaf_nodes(node):
    if node.value is not None:
        return 1
    return count_leaf_nodes(node.left) + count_leaf_nodes(node.right)

leaf_count = count_leaf_nodes(model.tree)
print("Number of leaf nodes: ", leaf_count)
```
