### Problem Statement 1 :

Write a Python program using NumPy to analyze the grades of multiple students across different subjects.

Your program should:

Create a list of 5 student names and a list of 3 subjects.

Generate a 5×3 matrix of random integer grades between 60 and 100 using np.random.randint().

Display the grades matrix, student names, and subject names.

Compute and display:

The average grade for each student (row-wise mean).

The average grade for each subject (column-wise mean).

Print each student’s name along with their average grade (rounded to 1 decimal place).

Determine and print the top-performing student and their average score.


#### Step 1: Import NumPy and define student and subject lists

```python
import numpy as np

student = ["Aditya","Vedant","Prapti","Sahili","Aniket"]
subject = ["Math","Science","English"]
```

#### Step 2: Generate a 5×3 matrix of random grades between 60 and 100

```python
grades = np.random.randint(60,101,size=(5,3))
print(grades)
```

#### Step 3: Display the grades matrix, student names, and subject names

```python
print("Subjects : ",subject)
print("Students : ",student)
print("Grades Matrix:\n :",grades)
```

#### Step 4: Compute average grade for each student (row-wise mean)

```python
student_average = np.mean(grades,axis = 1)
#axis=1 means we compute the mean across columns (i.e., per student).
print("Average grade per student:",student_average)
```

#### Step 5: Compute average grade for each subject (column-wise mean)

```python
subject_average = np.mean(grades,axis = 0)
# axis=0 means we compute the mean across rows (i.e., per subject).
print("Average grade per subject: ",subject_average)
```

#### Step 6: Print each student’s name with their average grade (rounded)

```python
print("\nStudent Averages:")
for name, avg in zip(student, student_average):
    print(f"{name} : {round(avg,1)}")
```

#### Step 7: Determine and print the top-performing student

```python
top_index = np.argmax(student_average)
top_student = student[top_index]
top_score = round(student_average[top_index],1)
# np.argmax() finds the index of the highest average.
# We use that index to get the student name and score.
print(F"Top-performing student : {top_student} with average score {top_score}")
```

### Problem Statement 2:

Write a Python program using NumPy to perform analysis on sales data of different products over six months.

Your program should:

Create a list of 6 months (Jan–Jun) and 3 products (Product A, Product B, Product C).

Generate a 6×3 matrix of random sales figures (between 1000 and 5000) using np.random.randint().

Display the sales matrix clearly (Months × Products).

Calculate and display:

Total monthly sales (sum across products).

Total sales per product (sum across months).

Identify and display:

The best-performing month (highest total sales).

The best-selling product (highest cumulative sales).

#### Step 1: Import NumPy and define months and products

```python
import numpy as np

months = ["Jan", "Feb", "Mar", "Apr", "May","June"]
products = ["Product A","Product B","Product C"]
```

#### Step 2: Generate a 6×3 matrix of random sales figures between 1000 and 5000

```python
sales = np.random.randint(1000,5001,size = (6, 3))
print(sales)
```

#### Step 3: Display the sales matrix clearly

```python
print("Monts: ",months)
print("product: ", products)
print("\nSales Matrix (Months x products):\n", sales)
```

#### Step 4: Calculate total monthly sales (sum across products)

```python
monthly_totals = np.sum(sales,axis = 1)
print("Total Monthly Sales:\n")
for month, total in zip(months,monthly_totals):
    print(f"{month} : {total}")
```

#### Step 5: Calculate total sales per product (sum across months)

```python
product_totals = np.sum(sales, axis = 0)
print("\nTotal sales per product: ")
for product, total in zip(products, product_totals):
    print(f"{product} : {total}")
```

#### Step 6: Identify the best-performing month

```python
best_month_index = np.argmax(monthly_totals)
best_month = months[best_month_index]
best_month_sales = month_totals[best_month_index]

print(f"Best-performing Month: {best_month} with sales of {best_month_sales}")
```

#### Step 7: Identify the best-selling product

```python
best_product_index = np.argmax(product_totals)
best_product = products[best_product_index]
best_product_sales = product_totals[best_product_index]

print(f"Best-Selling product: {best_product} with total sales of {best_product_sales}")
```
