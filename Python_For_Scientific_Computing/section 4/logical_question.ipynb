List Sum & Average:
Given a list of numbers, calculate the sum and average without using sum() or len() (use loops).

```python
num = input("Enter the number of elements you eants")
number_in_list = list(map(int, num.split()))
sum = 0
count = 0
for i in number_in_list:
    sum += i
    count += 1

print(sum)
print(f"{sum/count}")
```

### Combination of string and Dict

Word Frequency:
Take a string from the user and count how many times each word appears. Store results in a dictionary.

Example:
Input: "the cat and the dog"
Output: {'the': 2, 'cat': 1, 'and': 1, 'dog': 1}


```python
#Method 1: Using a for loop
text = "the cat and the dog"
word_freq = {}
for word in text.split():
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

print(word_freq)
```

```python
#Method 2: Using dict.get() for cleaner code
word_freq = {}
for word in text.split():
    word_freq[word] = word_freq.get(word, 0) + 1

print(word_freq)
```

```python
#Method 3: Dictionary Comprehension with set()
text = "the cat and the dog"
word_freq = {word: text.split().count(word) for word in text.split()}
print(word_freq)
```

## Leet code

#### 1> Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

*Example 1:*

Input: nums = [2,7,11,15,4,5], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]
Example 3:

Input: nums = [3,3], target = 6
Output: [0,1]

```python
nums = [2,7,11,15,4,5]
n = len(nums)
target = 9
for i in range(0,n-1):
    for j in range(1, n):
        if target == nums[i] + nums[j]:
            print([i,j])
```

```python
def two_sum(nums, target):
    n = len(nums)
    for i in range(0,n - 1):
        for j in range(1, n):
            if nums[i] + nums[j] == target:
                return [i, j]

result = two_sum([2, 7, 11, 15, 4, 5], 9)
print(result)
```
