### 2. NumPy – 2.5 Marks  
You are given a 2D NumPy array with missing values represented as `np.nan`. Replace all `np.nan` values with the **mean of their respective column** (computed ignoring NaN values). Use NumPy functions; **do not use explicit Python loops or Pandas**.

**Note:** You may assume at least one valid (non-NaN) value exists per column.

**Example:**  
Input:  
```python
[[1, np.nan, 3],  
 [4, 5, np.nan],  
 [7, 8, 9]]

````python
import numpy as np

arr = np.array([[1, np.nan, 3],  
               [4, 5, np.nan],  
               [7, 8, 9]])

#compute column means ignoring NaNs
col_mean = np.nanmean(arr, axis = 0)

#Replace NaNs wih column means
inds = np.where(np.isnan(arr))
arr[inds] = np.take(col_mean, inds[1])

print(arr)
````

### 3. Minimum Window Subsequence – 5 Marks  
Given two strings `s1` and `s2`, find the **minimum-length substring in `s1`** that contains **all characters of `s2` as a subsequence** (not necessarily consecutive).

A subsequence means characters appear in the same order but not necessarily adjacent.  
For example, `'ace'` is a subsequence of `'abcde'` (positions 0, 2, 4).

**Example:**  
Input: `s1 = 'abcdebde'`, `s2 = 'bde'`  
Output: `'bde'`  
Explanation:  
- `'b'` at position 2  
- `'d'` at position 3  
- `'e'` at position 5  
This forms the subsequence `'bde'`. Other windows like `'bdde'` are longer.

**Note:**  
- If multiple windows have the same minimum length, return the leftmost one.  
- If no such window exists, return an empty string.

````python
def min_window_subsequence(s1, s2):
    min_len = float('inf')  # track shortest length
    result = ""             # store final result

    for i in range(len(s1)):
        if s1[i] == s2[0]:  # Start match only if first char of s2 found
            j, k = i, 0     # j scans s1, k scans s2

            # Forward scan to match entire s2
            while j < len(s1):
                if s1[j] == s2[k]:
                    k += 1
                    if k == len(s2):  # Full match found
                        end = j

                        # backtrack to find start of minimal window
                        k -= 1
                        while j >= i:
                            if s1[j] == s2[k]:
                                k -= 1
                                if k < 0:
                                    break
                            j -= 1
                        start = j 

                        # update result if window is smaller
                        if end - start + 1 < min_len:
                            min_len = end - start + 1
                            result = s1[start:end+1]
                        break
                j += 1
    return result

print(min_window_subsequence("abcdebde", "bde"))
````

### 4. Palindrome Pairs – 5 Marks  
Given a list of words, find all pairs of distinct indices `(i, j)` where `i ≠ j`, such that concatenating `words[i] + words[j]` forms a palindrome.

**Example 1:**  
Input: `['bat', 'tab', 'cat']`  
Output: `[[0, 1], [1, 0]]`  
Explanation:  
- `'bat' + 'tab'` = `'battab'` → palindrome  
- `'tab' + 'bat'` = `'tabbat'` → palindrome  
- `'cat'` doesn't form palindromes with others.

**Example 2:**  
Input: `['a', 'abc', 'cba', '']`  
Output: `[[0, 3], [3, 0], [1, 2], [2, 1]]`  
Explanation:  
- `'a' + ''` = `'a'`, `'' + 'a'` = `'a'`  
- `'abc' + 'cba'` = `'abccba'` → palindrome

````python
def is_palindrome(word):
    return word == word[::-1]

def palindrome_pairs(words):
    result = []
    for i in range(len(words)):
        for j in range(len(words)):
            if i != j and is_palindrome(words[i] + words[j]):
                result.append([i,j])
    return result

print(palindrome_pairs(['bat', 'tab', 'cat']))
print(palindrome_pairs(['a', 'abc', 'cba', '']))
````

### 5. Letter Combinations of a Phone Number – 5 Marks  
Given a string containing digits from `2–9`, return all possible letter combinations that the number could represent based on the phone keypad mapping.

**Example:**  
Input: `'23'`  
Output: `['ad', 'ae', 'af', 'bd', 'be', 'bf', 'cd', 'ce', 'cf']`


````python
def letter_combinations(digits):
    if not digits:
        return []

    mapping = {
        '2': 'abc', '3': 'def', '4': 'ghi', '5': 'jkl',
        '6': 'mno', '7': 'pqrs', '8': 'tuv', '9': 'wxyz'
    }

    result = ['']
    for digit in digits:
        temp = []
        for prefix in result:
            for char in mapping[digit]:
                temp.append(prefix + char)
        result = temp
    return result

print(letter_combinations("23"))
````
