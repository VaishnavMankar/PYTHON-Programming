```python
df = pd.read_csv('bmw_with_missing_data.csv')
df.head()
```

#### print("1. Find how many missing values are in each column")

```python
print(df.isnull().sum())
```

#### print("2. Remove all rows with missing prices") 

```python
df1 = df.dropna(subset=['price'])
print(df1.isnull().sum())
```

#### print("3. Fill missing years with the most common year")

```python
most_common_year = df['year'].mode()[0]
df['year'].fillna(most_common_year, inplace = True)
```

#### print("4. Find and remove duplicate entries")

```python
df2 = df.drop_duplicates()
```

#### print("5. Calculate average price after cleaning")

```python
print(df['price'].dtype)
```

```python
df['price'] = pd.to_numeric(df['price'],errors = 'coerce')
```

```python
df = df.dropna(subset=['price'])
```

```python
average_price = df['price'].mean()
print(f"Average Price: {round(average_price,2)}")
```
