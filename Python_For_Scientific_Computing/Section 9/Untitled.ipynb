### Problem 2.1: Handle Missing Data
*Task*: Create a dataset with missing values and practice different cleaning strategies.
1. Remove all rows with ANY missing values
2. Remove rows only if ALL values are missing
3. Fill missing Math_Score with the average
4. Fill missing Science_Score with 0
5. Fill missing names with "Unknown"


```python
import numpy as np

data = {
    'name':['Aditya','Vadent',None,'Sahili','Prapti',None],
    'Math_score' : [85,np.nan,78,92,np.nan,87],
    'Science_Score':[90,85,np.nan,np.nan,70,np.nan]
}

df = pd.DataFrame(data)
print("Original Dataframe\n",df)
```

```python
# 1. Remove all rows with ANY missing values
df_any_missing_removed = df.dropna()
print("Row with any missing value removed:\n",df_any_missing_removed)
```

```python
# 2. Remove rows only if ALL values are missing
df_all_missing_remove = df.dropna(how='all')
print("Row with ALL values are missing remove:\n", df_all_missing_remove)
```

```python
# 3. Fill missing Math_Score with the average
math_avg = df['Math_score'].mean(skipna = True)
df['Math_score'].fillna(math_avg,inplace = True)
print('\nFilled missing Math_Score with average: \n', df)
```

```python
# 4. Fill missing Science_Score with 0
df['Science_Score'].fillna(0, inplace=True)
print("Filled missing Science_score with 0: \n",df)
```

```python
# 5. Fill missing names with "Unknown"
df['name'].fillna("Unknown",inplace=True)
print("Filling missing Name with Unknown:\n",df)
```

### Problem 3.1: Complex Filtering
*Task*: Using the BMW dataset, find cars that meet specific criteria.

*Your Tasks:*
1. Find all cars from 2016 or later
2. Find cars with price between 15000 and 25000
3. Find diesel cars with mileage less than 40000
4. Find the cheapest car for each fuel type



```python
df = pd.read_csv('bmw_with_missing_data.csv')
print("\nInformation about table\n",df.info())
print("\nNull values in tabel\n",df.isnull().sum())
```

```python
#Convert key columns to proper types, coercing errors to NaN:
df['year'] = pd.to_numeric(df['year'],errors = 'coerce')
df['price'] = pd.to_numeric(df['price'], errors = 'coerce')
df['mileage'] = pd.to_numeric(df['mileage'], errors = 'coerce')
```

```python
#Drop rows with missing values in critical columns
df_clean = df.dropna(subset=['year','price','mileage','fuelType'])
```

```python
# 1. Find all cars from 2016 or later
car_2016_plus = df_clean[df_clean['year'] >= 2016]
print(car_2016_plus.head())
```

```python
# 2. Find cars with price between 15000 and 25000
cars_in_price_range = df_clean[(df_clean['price'] >= 15000) & (df_clean['price'] <= 25000)]
print(cars_in_price_range.head())
```

```python
# 3. Find diesel cars with mileage less than 40000
diesel_low_mileage = df_clean[(df_clean['fuelType'].str.lower() == 'disel') & (df_clean['mileage'] < 40000)]
print(diesel_low_mileage)
```

```python
# 4. Find the cheapest car for each fuel type
cheapest_by_fuel = df_clean.loc[df_clean.groupby('fuelType')['price'].idxmin()]
print(cheapest_by_fuel)
# groupby('fuelType'): Groups the DataFrame by each fuel type (e.g., Diesel, Petrol, Electric).
# ['price'].idxmin(): For each group, finds the index of the row with the lowest price.
# df_clean.loc[...]: Retrieves the full row at that index — giving you the cheapest car for each fuel type.
```

### Problem 3.2: Grouping and Aggregation
*Task*: Analyze the BMW dataset by groups.

*Your Tasks:*
1. Group by fuel type and find average price
2. Group by year and count how many cars
3. Group by transmission type and find max, min, and average mileage
4. Find the most expensive car in each model category

```python
# 1. Group by fuel type and find average price
avg_price_by_fuel = df_clean.groupby('fuelType')['price'].mean()
print("\nAverage price by fuel type", avg_price_by_fuel)
```

```python
# 2. Group by year and count how many cars
car_count_by_year = df_clean.groupby('year')['model'].count()
print("\nNumber fo car by year\n",car_count_by_year)
```

```python
# 3. Group by transmission type and find max, min, and average mileage
mileage_stats_by_transmission = df_clean.groupby('transmission')['mileage'].agg(['max','min','mean'])
print("\nMileage stats by transmission type:\n",mileage_stats_by_transmission)
```

```python
# 4. Find the most expensive car in each model category
most_expensive_by_model = df_clean.loc[df_clean.groupby('model')['price'].idxmax()]
print("\nMost expensive car by model:\n", most_expensive_by_model[['model','price']])
```
